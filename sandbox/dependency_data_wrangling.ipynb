## Pt 1: creating a dependency report using @segmentio/dependency-report
#### (https://github.com/segmentio/dependency-report) 

## First, follow installation in the README of the package
If installation is being difficult, try `sudo npm install -g @segment/dependency-report`

## Using the CLI
the syntax is `dependency-report ['file-path'] --packages='dependencies-you-are-parsing-for'`
for example, if I wanted to look for instances of our UI kit use in the topmost components directory, I'd command:
`dependency-report './components/**/*.*(js|jsx)' --packages=@washingtonpost/wpds-ui-kit`

## Data wrangling
I am going to start with Python because I am more familiar with it. Here, I find how many times each imported component is used in the topmost components directory in the wpds-ui-kit repo.
I will visualize which component is most used and how many times each is used.

```python
counts: dict = {}

for comp in data["filepathsForExports"]:
    counts[comp] = len(data["filepathsForExports"][comp])

print(counts)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize =(10,7))
plt.pie(counts.values(), labels = counts.keys(), autopct='%1.1f%%')
plt.title("Components and their relative usage (in wpds-docs/components)")
plt.show()
```

```python
y = list(counts.keys())
x = list(counts.values())

plt.barh(y, x)
plt.title("Instances of each component (in wpds-docs/components)")
plt.show()
```

Now, I want to find out which file uses the most components (stripped parent directory for clarity)

```python
count: dict = {}
listing: dict = data["filepathsForExports"]
for component in listing:
    for path in listing[component]:
        if path.removeprefix("./components/") in count:
            count[path.removeprefix("./components/")] += 1
        else:
            count[path.removeprefix("./components/")] = 1

print(count)
```

```python
y = list(count.keys())
x = list(count.values())

plt.barh(y, x)
plt.title("Number of components each file in wpds-docs/components imports")
plt.show()
```

## Pt. 2: using Github's search API

Github's search api (https://docs.github.com/en/rest/search#about-the-search-api) offers the ability to search for specific data across the github universe. I want to find which repositories under the washington post's perview install react as a dependency and which install our ui kit. Below is a simple program that will search for just that. It is interactive, but you can hard-code it to take whatever inputs you want. I suggest 'WPMedia' or 'washingtonpost' for org, 'react' or 'wpds-ui-kit' for code, and 'package.json' for file. I've accounted for duplicate installations in the same repository, and my authorization token is hard-coded (oops).

```python
import requests

c: bool = True  # marks continuation of the loop
seen: list = []

while c:
    org = input("\nWhich org would you like to search? ")
    code = input("Which code would you like to match? ")
    filename = input("Which file would you like to search? ")

    url = f"https://api.github.com/search/code?q={code}+user:{org}+filename:{filename}"

    response = requests.get(url=url, headers={'Authorization': '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'})
    print(f"\nThe status code of this request is: {response.status_code}\n")

    json_response = response.json()
    response.close()
    items = json_response["items"]

    count: int = 0
    # traverses the response for repository names and prints them out to the user
    for item in items:
        if item['repository']['name'] not in seen:
            print(f"Repository {count} using '{code}' in org '{org}': {item['repository']['name']}")
            count += 1
            seen.append(item['repository']['name'])
    if count == 0:
        print(f"NO repositories found using '{code}' in org '{org}'")
    
    c2 = input("\nWould you like to continue? ")
    if c2.lower() == "no":
        c = False
    seen = []
```
